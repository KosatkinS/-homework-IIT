## <font size="5" color="black" face="Times"> <b>3.1.Разбор теоретических вопросов</font><a name = 'part31'>

<font size="3" color="black" face="Times"><div align="justify">
1. **Software versioning** — это процесс создания уникальных имен или номеров для различных версий продуктов программного обеспечения.
2. **Git** — распределённая система управления версиями, созданная Линусом Торвальдсом для управления разработкой ядра Linux.<br>
Основные задачи Git:<br>
хранить информацию обо всех изменениях в коде, начиная с самой первой строчки;<br>
обеспечение удобства командной работы над кодом.<br>
В Git каждая рабочая копия кода сама по себе является репозиторием. Это позволяет всем разработчикам хранить историю изменений в полном объёме.
3. **Локальный репозиторий** — это репозиторий, который хранится на вашем локальном компьютере. В контексте систем управления версиями, таких как Git, локальный репозиторий представляет собой место, где сохраняются все файлы и история изменений вашего проекта, а также метаданные, такие как коммиты, ветки и теги.
